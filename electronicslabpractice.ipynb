```python
import numpy as np 
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn import metrics
import pickle
dataFrame= pd.read_csv("../input/maternal-health-risk-data/Maternal Health Risk Data Set.csv")
dataFrame["RiskLevel"].replace({"high risk": 2, "mid risk": 1,"low risk":0}, inplace=True)
```

```python
Xcols=dataFrame.columns[0:-1]
X=dataFrame[Xcols]
ycols=dataFrame.columns[-1]
y=dataFrame[ycols]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X_train.values,y_train.values)
y_pred=clf.predict(X_test.values)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
p=[[30,140,85,7.0,98.0,70]]
predAns=clf.predict(p)
predAns
```

```python
filename = 'finalized_model.pkl'
pickle.dump(clf, open(filename, 'wb'))
```
